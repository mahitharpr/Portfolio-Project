```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://en.wikipedia.org/wiki/Economy_of_India'
```

```python
page= requests.get(url)
page
```

```python
soup = BeautifulSoup (page.text, 'html')
soup
```

```python
table=soup.find('table', class_='wikitable')
print(table)
```

```python
titles = table.find_all('th')
```

```python
all_titles = [ title.text.strip() for title in titles]
all_titles
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns = all_titles)
df
```

```python
data = table.find_all('tr')
data
```

```python
for row in data[1:]:
    row_data = row.find_all('td')
    all_row_data = [ row.text.strip() for row in row_data]
    print(all_row_data)
```

```python
for row in data[1:]:
    row_data = row.find_all('td')
    all_row_data = [ row.text.strip() for row in row_data]

    length = len(df)
    df.loc[length]=all_row_data
```
